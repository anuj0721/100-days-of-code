```python
n = int(input("How many records?: "))
Name = []
RegNo = []
PhysicsMarks = []
ChemistryMarks = []
MathMarks = []
TwelthPercentage = []
Course = []
Flag = []
for i in range(n):
    studentname = str(input("Enter student Name: "))
    Name.append(studentname)
    studentregno = input("Enter Registration Number: ")
    RegNo.append(studentregno)
    physicsmarks = int(input("Enter Physics Marks: "))
    PhysicsMarks.append(physicsmarks)
    chemistrymarks = int(input("Enter Chemistry Marks: "))
    ChemistryMarks.append(chemistrymarks)
    mathsmarks = int(input("Enter Maths Marks: "))
    MathMarks.append(mathsmarks)
    intermediateper = int(input("Enter 12th Percentage: "))
    TwelthPercentage.append(intermediateper)
    courseopted = str(input("Enter Course Opted: "))
    Course.append(courseopted)
    flag = str(input("Enter status as Active/Inactive: "))
    Flag.append(flag)

data = [Name,RegNo,PhysicsMarks,ChemistryMarks,MathMarks,TwelthPercentage,Course,Flag] 
import pandas as pd
df=pd.DataFrame(data, index=['Name','RegNo','Physics Marks','Chemistry Marks','Math Marks','12Th Percentage','Course Opted','Flag'])
```

```python
# Transpose the dataframe
dft = (df.T)
```

```python
# Store Data in csv file
dft.to_csv("Student_record.csv")
```
